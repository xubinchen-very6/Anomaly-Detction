```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
raw_path = '../../nyc_taxi.csv'
train_path = '../../nyc_taxi_train.csv'
test_path = '../../nyc_taxi_test.csv'
```

```python
data = pd.read_csv(raw_path,usecols=[1,2])
train = pd.read_csv(train_path,usecols=[2,3])
test = pd.read_csv(test_path,usecols=[2,3])
```

```python
lstmEncDec_path = '../anomaly_detection/lstmEncDnc_yHat1.npy'
lstmEncDec = np.load(lstmEncDec_path)
```

```python
y_true = test.values[101::,1]
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit_transform(data['value'].values.reshape(-1,1))
```

```python
lstmEncDec_inv = scaler.inverse_transform(lstmEncDec)
```

```python
from sklearn.metrics import r2_score
```

```python
plt.figure(figsize=(40,10))
plt.plot(lstmEncDec.reshape(-1),label='pred')
plt.plot(y_true,label = 'ground_truth')
plt.legend()
plt.show()
print(r2_score(y_true,lstmEncDec_inv))
```
